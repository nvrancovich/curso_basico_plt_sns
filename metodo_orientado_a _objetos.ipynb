```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 5, 11)
y = x**2
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.5, 0.5])

axes.plot(x, y, 'b')
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 1, 1])
axes2 = fig.add_axes([0.3, 0.6, 0.3, 0.3])

axes.plot(x, y, 'b')
axes2.plot(y, x, 'r')
```
